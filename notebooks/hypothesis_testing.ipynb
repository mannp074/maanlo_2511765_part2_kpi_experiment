```python
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest

# Load cleaned data
df = pd.read_excel("experiment_analysis.xlsx", sheet_name="cleaned_data")

# Split groups
control = df[df['experiment_group'] == 'Control']
treatment = df[df['experiment_group'] == 'Treatment']

# Success counts (paid conversions)
control_success = control['converted_to_paid'].sum()
treatment_success = treatment['converted_to_paid'].sum()

# Sample sizes
control_n = len(control)
treatment_n = len(treatment)

# Perform one-tailed two-proportion z-test
successes = [treatment_success, control_success]
sample_sizes = [treatment_n, control_n]

z_stat, p_value = proportions_ztest(
    count=successes,
    nobs=sample_sizes,
    alternative='larger'
)

# Display results
print("="*60)
print("HYPOTHESIS TEST RESULTS")
print("="*60)

print(f"Treatment Conversions : {treatment_success}")
print(f"Treatment Users       : {treatment_n}")

print(f"Control Conversions   : {control_success}")
print(f"Control Users         : {control_n}")

print("\nZ Statistic :", round(z_stat,4))
print("P-value     :", round(p_value,6))

alpha = 0.05

print("\nDecision Rule:")
print("Reject H0 if p-value < 0.05")

if p_value < alpha:
    print("\nDecision: Reject H0")
    print("Conclusion: Treatment significantly improves paid conversion rate.")
else:
    print("\nDecision: Fail to Reject H0")
    print("Conclusion: No statistically significant improvement detected.")
    
```
