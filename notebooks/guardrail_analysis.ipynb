```python
import pandas as pd

# Load cleaned data
df = pd.read_excel("experiment_analysis.xlsx", sheet_name="cleaned_data")

# Split groups
control = df[df['experiment_group'] == 'Control']
treatment = df[df['experiment_group'] == 'Treatment']

# Guardrail metrics
guardrail_summary = pd.DataFrame({
    "Metric": [
        "Refund Rate",
        "Support Ticket Rate",
        "Average Days to Convert",
        "Average Engagement Score",
        "Average Revenue per Converted User"
    ],
    "Control": [
        control['refund_requested'].mean(),
        (control['support_tickets_30d'] > 0).mean(),
        control.loc[control['converted_to_paid']==1, 'days_to_convert'].mean(),
        control['engagement_score'].mean(),
        control.loc[control['converted_to_paid']==1, 'revenue_30d'].mean()
    ],
    "Treatment": [
        treatment['refund_requested'].mean(),
        (treatment['support_tickets_30d'] > 0).mean(),
        treatment.loc[treatment['converted_to_paid']==1, 'days_to_convert'].mean(),
        treatment['engagement_score'].mean(),
        treatment.loc[treatment['converted_to_paid']==1, 'revenue_30d'].mean()
    ]
})

print("="*60)
print("GUARDRAIL METRICS")
print("="*60)

print(guardrail_summary)

# Segment-level conversion analysis
segment_risk = df.groupby(['region', 'experiment_group'])['converted_to_paid'].mean().unstack()

print("\n" + "="*60)
print("REGION-LEVEL CONVERSION RATES")
print("="*60)

print(segment_risk)
```
